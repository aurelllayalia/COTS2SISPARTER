```python
#Import library os untuk melakukan perintah ping
import os

#Import library re untuk melakukan filtering/regex
import re

#Import library time untuk mengukur waktu
import time

#Mengambil waktu awal menggunakan time.time()
start = time.time()

#Menggunakan regex filter r"Received = (\d)
#Regex ini digunakan untuk mencari kata "Received = x" dimana x adalah angka (\d)
received_packages = re.compile(r"Received = (\d)")

#Melihat 3 baris paling bawah dari hasil ping
#Terdapat baris: Packets: Sent = 2, Received = 2, Lost = 0 (0% loss),
#Yang dilihat adalah jumlah Received. Kebetulan hasilnya adalah 2
#Pada ping x.x.x.x -n 2 artinya dilakukan ping sebanyak 2 kali
#Jika received = 0 maka host mati
#Jika received = 1 maka host hidup tetapi ada ping yang hilang
#Jika received = 2 maka host hidup
status = ("tidak ada respon","hidup tetapi ada yang hilang","hidup")

#Melakukan ping untuk 20 host
for suffix in range(1,20):
    #Menentukan IP yang akan diping, hanya akan diubah bagian akhirnya
    ip = "10.30.32."+str(suffix)
    
    #Melakukan ping -n 2
    ping_out = os.popen("ping -n 2 "+ip,"r")
    print(("... pinging ",ip))

    #Loop forever
    while True:
        #Membaca hasil dari perintah ping per line
        line = ping_out.readline()
        
        #Akan keluar dari loop jika tidak ada lagi line
        if not line: break
        
        #Melakukan filtering dengan regex. gunakan findall dan regex filter yang sudah ditentukan
        n_received = received_packages.findall(line)
        
        #Menampilkan hasil regex untuk mengetahui banyaknya respon ping yang diperoleh 
        if n_received:
            print((ip + ": " + status[int(n_received[0])]))

#Mengambil waktu berakhir
end = time.time()

#Menampilkan selisih waktu akhir dengan awal untuk mengetahui lama waktu eksekusi
print((end-start))
```
