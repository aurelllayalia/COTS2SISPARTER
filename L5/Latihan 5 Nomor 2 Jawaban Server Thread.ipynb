```python
#Import socket, sys, traceback dan threading
import socket
import sys
import traceback
import threading

#Menjalankan server
def main():
    start_server()

#Fungsi saat server dijalankan
def start_server():
    #Menentukan IP server
    ip_server = 'localhost'
    
    #Menentukan port server
    post_server = 33000

    #Membuat socket bertipe TCP
    soc = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    
    #Option socket
    soc.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
    print("Socket dibuat")

    #Melakukan bind
    try:
        soc.bind((ip_server, post_server))
    except:
        #Exit pada saat error
        print("Bind gagal. Error : " + str(sys.exc_info()))
        sys.exit()

    #Listen hingga 5 antrian
    soc.listen(5)
    print("Socket mendengarkan")

    #Infinite loop, jangan reset setiap ada request
    while True:
        #Menerima koneksi
        conn, addr = soc.accept()
        
        #Mendapatkan IP dan port
        ip = str(addr[0])
        port = str(addr[1])
        print("Connected dengan " + ip + ":" + port)

        #Menjalankan thread untuk setiap koneksi yang terhubung
        try:
            threading.Thread(target=client_thread, args=(conn, ip, port)).start()
        except:
            #Print kesalahan jika thread tidak berhasil dijalankan
            print("Thread tidak berjalan")
            traceback.print_exc()

    #Menutup socket
    soc.close()

def client_thread(connection, ip, port, max_buffer_size = 4096):
    #Flag koneksi
    is_active = True

    #Selama koneksi aktif
    while is_active:

        #Menerima pesan dari client
        client_input = connection.recv(max_buffer_size)        
        
        #Mendapatkan ukuran pesan
        client_input_size = sys.getsizeof(client_input)
        
        #Print jika pesan terlalu besar
        if client_input_size > max_buffer_size:
            print("The input size is greater than expected {}")

        #Mendapatkan pesan setelah didecode
        client_input = client_input.decode('utf8').rstrip()
        
        #Jika "quit" maka flag koneksi = false, matikan koneksi
        if "quit" in client_input:
            #Mengubah flag
            is_active = False
            print("Client meminta keluar")
            
            #Mematikan koneksi
            connection.close()
            print("Connection " + ip + ":" + port + " ditutup")
            
        else:
            #Menampilkan pesan dari client
            print(client_input)           
            
#Memanggil fungsi utama
if __name__ == "__main__":
    main()
```
