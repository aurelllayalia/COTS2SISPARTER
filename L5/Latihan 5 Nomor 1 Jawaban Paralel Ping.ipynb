```python
#Import os, re dan threading
import os
import re
import threading

#Import time
import time

#Membuat kelas ip_check
class ip_check(threading.Thread):
    
    #Fungsi __init__; init untuk assign IP dan hasil respons = -1
    def __init__ (self, ip):
        threading.Thread.__init__(self)
        self.ip = ip
        self.successful_pings = -1
        
    #Fungsi utama yang diekseskusi ketika thread berjalan
    def run(self):
        #Melakukan ping dengan perintah ping -n (gunakan os.popen())
        ping_out= os.popen('ping -n 2 ' + ip,'r')
        
        #Loop forever
        while True:
            #Membaca hasil respon setiap baris
            line = ping_out.readline()
            
            #Break jika tidak ada line lagi
            if not line : break
            
            #Membaca hasil per line dan temukan pola Received = x
            n_received = re.findall(received_packages, line)
            
            #Menampilkan hasilnya
            if n_received:
                self.successful_pings = int(n_received[0])
                
    #Fungsi untuk mengetahui status; 0 = tidak ada respon, 1 = hidup tapi ada loss, 2 = hidup
    def status(self):
        #0 = tidak ada respon
        if self.successful_pings == 0:
            return 'Tidak ada respon'
        #1 = ada loss
        elif self.successful_pings == 1: 
            return 'Hidup tapi ada loss'
        #2 = hidup
        elif self.successful_pings== 2:
            return 'Hidup'
        #-1 = seharusnya tidak terjadi
        else: 
            return 'Seharusnya tidak terjadi'
        
#Membuat regex untuk mengetahui isi dari r"Received = (\d)"
received_packages = re.compile(r'Received = (\d)')

#Mencatat waktu awal
start = time.time()

#Membuat list untuk menampung hasil pengecekan
check_results = []

#Melakukan ping untuk 20 host
for suffix in range(1,20):
    #Menentukan IP host apa saja yang akan di ping
    ip = '192.168.100.'+str(suffix)
    
    #Memanggil thread untuk setiap IP
    current = ip_check(ip)
    
    #Memasukkan setiap IP dalam list
    check_results.append(current)
    
    #Menjalankan thread
    current.start()

#Untuk setiap IP yang ada di list
for el in check_results:
    
    #Menunggu hingga thread selesai
    el.join()
    
    #Mendapatkan hasilnya
    print('Status', el.ip, el.status())

#Mencatat waktu berakhir
end = time.time()

#Menampilkan selisih waktu akhir dan awal
print('\nSelisih waktu : ',end+start)
```
